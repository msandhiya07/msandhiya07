# Histogram     Bar chart

```python
from matplotlib import pyplot as plt
```

```python
marks=[90,30,40,60,35,50,80]
grade_intervals=[0,35,70,100]
plt.title("STUDENT GRADE")
plt.hist(marks,grade_intervals,histtype="stepfilled",rwidth=0.7,facecolor="grey")    #inly small letters for facecolor
plt.xticks([0,35,70,100])      #x axis values
plt.xlabel("Percentage")
plt.ylabel("NO of Students")
plt.show()
```

# pie chart

```python
from matplotlib import pyplot as plt
```

```python
student_performance=["Execellent","Good","Average","Poor"]
student_values=[15,25,12,8]
plt.pie(student_values)

plt.show()
```

```python
plt.pie(student_values,labels=student_performance)
plt.show()
```

```python
plt.pie(student_values,labels=student_performance,startangle=90)
```

```python
plt.pie(student_values,labels=student_performance,startangle=180,shadow=True,colors=["blue","red","orange","black"])
```
